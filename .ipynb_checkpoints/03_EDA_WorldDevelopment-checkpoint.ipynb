## Selected Problem Statement:
__Targeted Funding for "Hollow Growth" Regions: Top-level economic growth often masks underlying crises, such as rapid population expansion diluting wealth or rising incomes failing to improve life expectancy. The Policy Steering Committee needs to analyze historical population, GNI per capita, and life expectancy trends (1960-2020) to identify nations trapped in these hollow growth cycles. Pinpointing these disconnects will allow the agency to accurately target a $100M intervention fund to countries suffering from stalled human development.__

## 1. Introduction & Objectives

### 1.1 Context & Background
For decades, GDP-style economic indicators like GNI per capita have been used as a shorthand for "development." But rising income does not automatically translate into a better quality of life. Two forces can quietly undermine that link:

1. **Population dilution** — total national wealth can grow while wealth *per person* stagnates, because the population is growing just as fast.
2. **Health stagnation** — income per person can rise while life expectancy — a core proxy for public health, healthcare access, and quality of life — fails to keep pace, or even declines (due to disease, conflict, or weak health systems).

We call countries experiencing either of these patterns **"hollow growth"** regions: places where the headline economic numbers look fine, but human development is stalling underneath. A Policy Steering Committee overseeing a $100M intervention fund needs a data-driven way to identify these countries so the money reaches the places that need it most, rather than the places that just look good on a GDP chart.

This notebook uses three Gapminder datasets — **population**, **GNI per capita (Atlas method, constant 2021 USD)**, and **life expectancy** — spanning 1800–2100, restricted to the **1960–2020** window specified in the problem statement, to identify these countries.

### 1.2 Core Analytical Objectives
* Clean and merge the three datasets into a single, tidy country-year panel for 1960-2020.
* Quantify each country's GNI per capita growth, life expectancy gains, and population growth over the full period.
* Distinguish real "hollow growth" from countries that simply started with high life expectancy and have less room left to improve (a ceiling effect).
* Identify a shortlist of countries where income grew strongly but life expectancy gains badly lagged expectations — the primary candidates for the $100M fund.
* Illustrate at least one concrete case study showing how a national shock (e.g., disease, conflict) can produce hollow growth even amid decent economic performance.

## 2. Environment Setup & Data Collection
*(Rubric Checkpoint: Ensure all modules are imported cleanly with appropriate standard aliases. Data files must be loaded using relative paths for replicability.)*

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

```

```python
# Load data using strictly relative file paths
pop_raw = pd.read_csv('../Data/population.csv')
gni_raw = pd.read_csv('../Data/gni_per_cap_atlas_method_con2021.csv')
life_raw = pd.read_csv('../Data/life_expectancy.csv')

print(f'Population:       {pop_raw.shape[0]} countries, {pop_raw.shape[1]-1} year columns')
print(f'GNI per capita:    {gni_raw.shape[0]} countries, {gni_raw.shape[1]-1} year columns')
print(f'Life expectancy:   {life_raw.shape[0]} countries, {life_raw.shape[1]-1} year columns')
```

---
## 3. Data Handling

```python
# Initial assessment: Check shapes, null values, and summary data types
for name, d in [('population', pop_raw), ('gni_per_capita', gni_raw), ('life_expectancy', life_raw)]:
    print(f'--- {name} ---')
    print('shape:', d.shape)
    print('dtypes (sample):', d.dtypes.value_counts().to_dict())
    print('total nulls:', d.isnull().sum().sum())
    print()
```

**Observations:** All three files are in "wide" Gapminder format — one row per country, one column per year. The `population` file stores values as strings with `k` / `M` / `B` suffixes (e.g. `27.2k`, `3.28M`), while `gni_per_capita` mixes plain numbers with the same `k`/`M` suffixes for very large economies. `life_expectancy` is already numeric. None of the files use a consistent country list — we'll need to align on the intersection when we merge.

```python
# Data cleaning operations (handling data types, text styling, missing values, etc.)

def to_number(x):
    """Convert Gapminder-style shorthand values ('3.28M', '27.2k', '1.4B') to floats."""
    if pd.isnull(x):
        return np.nan
    if isinstance(x, (int, float)):
        return float(x)
    x = str(x).strip()
    mult = 1
    if x.endswith('k'):
        mult, x = 1e3, x[:-1]
    elif x.endswith('M'):
        mult, x = 1e6, x[:-1]
    elif x.endswith('B'):
        mult, x = 1e9, x[:-1]
    try:
        return float(x) * mult
    except ValueError:
        return np.nan

YEARS = [str(y) for y in range(1960, 2021)]  # Restrict to the 1960-2020 window per the problem statement

def to_long(df, value_name):
    """Melt a wide Gapminder file (country x year columns) into tidy long format."""
    cols = ['country'] + [c for c in YEARS if c in df.columns]
    long_df = df[cols].melt(id_vars='country', var_name='year', value_name=value_name)
    long_df['year'] = long_df['year'].astype(int)
    return long_df

pop_long = to_long(pop_raw, 'population_raw')
pop_long['population'] = pop_long['population_raw'].apply(to_number)
pop_long = pop_long.drop(columns='population_raw')

gni_long = to_long(gni_raw, 'gni_per_capita_raw')
gni_long['gni_per_capita'] = gni_long['gni_per_capita_raw'].apply(to_number)
gni_long = gni_long.drop(columns='gni_per_capita_raw')

life_long = to_long(life_raw, 'life_expectancy')

# Inner-merge on country + year so we only keep countries present in ALL THREE datasets
df = (pop_long
      .merge(gni_long, on=['country', 'year'], how='inner')
      .merge(life_long, on=['country', 'year'], how='inner'))

print('Merged panel shape:', df.shape)
print('Countries retained:', df.country.nunique())
print('Nulls remaining:')
print(df.isnull().sum())
df = df.dropna(subset=['life_expectancy'])
print('Rows after dropping the handful of missing life-expectancy records:', len(df))
```

We lose a small number of micro-states / territories in the merge (e.g. Monaco, Taiwan, San Marino aren't present in the GNI file, and Holy See / Liechtenstein aren't in the life expectancy file). This is an acceptable tradeoff — these are tiny outlier economies that wouldn't be realistic targets for the $100M fund regardless.

## 4. Exploratory Data Analysis (EDA)
*(Rubric Checkpoint: Demonstrate mastery of basic descriptive statistics, sorting records, and applying Boolean masking techniques to isolate specific baseline metrics.)*

### 4.1. Descriptive Statistics for the Full 1960-2020 Panel

```python
# Compute appropriate summary statistics
df[['population', 'gni_per_capita', 'life_expectancy']].describe()
```

**Reading the table:** GNI per capita ranges from very low values to well over $100k, reflecting the huge spread between the poorest and richest economies. Life expectancy ranges from the low teens (wartime/epidemic years, see Section 4.4) up into the low 80s. Population is heavily right-skewed, as expected — a handful of countries (China, India, the US) dominate the mean.

### 4.2. Boolean Masking: Isolating Baseline (1960) Conditions

```python
# Boolean masking to isolate the 1960 baseline snapshot, and flag "poor + short-lived" starting conditions
baseline_1960 = df[df['year'] == 1960].copy()

# Boolean mask: countries that started 1960 both low-income AND low life expectancy
low_income_mask = baseline_1960['gni_per_capita'] < baseline_1960['gni_per_capita'].median()
low_life_mask = baseline_1960['life_expectancy'] < baseline_1960['life_expectancy'].median()

disadvantaged_1960 = baseline_1960[low_income_mask & low_life_mask]

print(f"Countries below the median on BOTH GNI per capita and life expectancy in 1960: {len(disadvantaged_1960)} of {len(baseline_1960)}")
disadvantaged_1960[['country', 'gni_per_capita', 'life_expectancy', 'population']].sort_values('gni_per_capita').head(10)
```

### 4.3. Sorting: Fastest / Slowest Population Growth (1960 vs 2020)

```python
# Sorting to find fastest-growing and slowest-growing populations
pop_1960 = baseline_1960.set_index('country')['population']
pop_2020 = df[df['year'] == 2020].set_index('country')['population']

pop_growth = pd.DataFrame({'population_1960': pop_1960, 'population_2020': pop_2020}).dropna()
pop_growth['pct_change'] = (pop_growth['population_2020'] / pop_growth['population_1960'] - 1) * 100

print("Fastest population growth 1960-2020:")
display(pop_growth.sort_values('pct_change', ascending=False).head(5))

print("Slowest / negative population growth 1960-2020:")
display(pop_growth.sort_values('pct_change', ascending=True).head(5))
```

Notably, small Gulf states (e.g. Qatar, UAE) dominate the fastest-growth list — driven by labor migration tied to oil wealth, not organic birth rates. This is important context for the hollow-growth analysis in Section 4.4: rapid population growth in these cases is a *symptom* of the same oil economy that inflates GNI per capita, rather than a dilution effect.

### 4.4. Target Analysis: Building the "Hollow Growth" Index

```python
# Target Analysis Block 1: quantify each country's growth over the full 1960-2020 window

start = df[df.year == 1960].set_index('country')
end = df[df.year == 2020].set_index('country')
common_countries = start.index.intersection(end.index)
start, end = start.loc[common_countries], end.loc[common_countries]

N_YEARS = 2020 - 1960

summary = pd.DataFrame({
    'gni_1960': start.gni_per_capita, 'gni_2020': end.gni_per_capita,
    'life_1960': start.life_expectancy, 'life_2020': end.life_expectancy,
    'pop_1960': start.population, 'pop_2020': end.population,
}).dropna()

# Compound annual growth rate (CAGR) for GNI per capita and population
summary['gni_cagr_pct'] = ((summary.gni_2020 / summary.gni_1960) ** (1 / N_YEARS) - 1) * 100
summary['pop_cagr_pct'] = ((summary.pop_2020 / summary.pop_1960) ** (1 / N_YEARS) - 1) * 100
summary['life_exp_change_years'] = summary.life_2020 - summary.life_1960

# Total national income (GNI per capita x population) isolates population-dilution effects:
# if total income grows much faster than per-capita income, population growth is "eating" the gains.
summary['total_gni_1960'] = summary.gni_1960 * summary.pop_1960
summary['total_gni_2020'] = summary.gni_2020 * summary.pop_2020
summary['total_gni_cagr_pct'] = ((summary.total_gni_2020 / summary.total_gni_1960) ** (1 / N_YEARS) - 1) * 100
summary['dilution_gap_pct'] = summary.total_gni_cagr_pct - summary.gni_cagr_pct

print(f"{len(summary)} countries have complete 1960 and 2020 data for all three metrics.")
summary[['gni_cagr_pct', 'pop_cagr_pct', 'life_exp_change_years', 'dilution_gap_pct']].describe()
```

**Controlling for the ceiling effect:** A naive flag (GNI growth above median AND life expectancy gain below median) unfairly penalizes countries like Ireland or Estonia — they started 1960 already near 70 years of life expectancy, so a smaller *absolute* gain is expected, not a sign of hollow growth. To fix this, we regress each country's life expectancy gain on its 1960 starting level, and flag countries whose *actual* gain falls well short of what that regression predicts — i.e., a large negative residual — while also having above-median GNI growth.

```python
# Regress life expectancy gain on the 1960 starting level to control for the ceiling effect
coeffs = np.polyfit(summary.life_1960, summary.life_exp_change_years, 1)
summary['expected_life_gain'] = np.polyval(coeffs, summary.life_1960)
summary['life_gain_residual'] = summary.life_exp_change_years - summary.expected_life_gain

gni_median = summary.gni_cagr_pct.median()
residual_median = summary.life_gain_residual.median()

summary['hollow_growth_flag'] = (summary.gni_cagr_pct > gni_median) & (summary.life_gain_residual < residual_median)

print(f"Countries flagged as 'hollow growth' candidates: {summary.hollow_growth_flag.sum()} of {len(summary)}")

hollow_growth_shortlist = (summary[summary.hollow_growth_flag]
                           .sort_values('life_gain_residual')
                           [['gni_1960', 'life_1960', 'gni_cagr_pct', 'life_exp_change_years', 'life_gain_residual', 'pop_cagr_pct']])
hollow_growth_shortlist.head(15)
```

**Key result:** Botswana and Eswatini top the shortlist — both had respectable GNI-per-capita growth (helped by diamond/sugar exports and southern African trade), but their life-expectancy residuals are deeply negative (roughly –9 to –10 years worse than their 1960 starting point would predict). This lines up with documented history: both countries were hit exceptionally hard by the HIV/AIDS epidemic from the late 1980s through the 2000s, which erased decades of health gains even as GDP kept climbing. This is exactly the kind of "hollow growth" the Committee should watch for — and it demonstrates the index is picking up real, not spurious, signal.

### 4.5. Visualizations

```python
# Visualization 1: GNI growth vs. life-expectancy-gain residual, with hollow growth candidates highlighted
fig, ax = plt.subplots(figsize=(9, 6))

colors = summary['hollow_growth_flag'].map({True: '#d62728', False: '#1f77b4'})
ax.scatter(summary['gni_cagr_pct'], summary['life_gain_residual'], c=colors, alpha=0.7, edgecolor='white', linewidth=0.5, s=45)

ax.axhline(residual_median, color='grey', linestyle='--', linewidth=1)
ax.axvline(gni_median, color='grey', linestyle='--', linewidth=1)

# Label the most extreme hollow-growth cases
for country in hollow_growth_shortlist.head(5).index:
    row = summary.loc[country]
    ax.annotate(country, (row.gni_cagr_pct, row.life_gain_residual), fontsize=9,
                xytext=(5, 5), textcoords='offset points')

ax.set_xlabel('GNI per capita CAGR, 1960-2020 (%)')
ax.set_ylabel('Life expectancy gain vs. expected, given 1960 starting point (years)')
ax.set_title('Identifying "Hollow Growth" Countries: Income Growth vs. Health Gains')
ax.legend(handles=[
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#d62728', markersize=8, label='Hollow growth candidate'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#1f77b4', markersize=8, label='Other')
], loc='lower right')
plt.tight_layout()
plt.savefig('../Presentation/viz1_hollow_growth_scatter.png', dpi=150)
plt.show()
```

**Visualization 1 Interpretation:**
> Each point is a country. The x-axis shows how fast GNI per capita grew (1960-2020); the y-axis shows how much life expectancy gained *relative to what we'd expect* given where the country started. Countries in the upper-right or lower-right quadrants had strong income growth; the ones highlighted in red (lower-right) are where that income growth did *not* translate into a proportional health payoff — the clearest candidates for the intervention fund, since money alone clearly isn't reaching health outcomes in these places.

```python
# Visualization 2: Case study - Botswana & Eswatini, GNI per capita vs. life expectancy over time
case_countries = ['Botswana', 'Eswatini']
fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharex=True)

for ax, country in zip(axes, case_countries):
    sub = df[(df.country == country)].sort_values('year')
    ax2 = ax.twinx()
    ax.plot(sub.year, sub.gni_per_capita, color='#2ca02c', linewidth=2, label='GNI per capita')
    ax2.plot(sub.year, sub.life_expectancy, color='#d62728', linewidth=2, label='Life expectancy')
    ax.set_title(country)
    ax.set_xlabel('Year')
    ax.set_ylabel('GNI per capita (constant 2021 USD)', color='#2ca02c')
    ax2.set_ylabel('Life expectancy (years)', color='#d62728')
    ax.tick_params(axis='y', labelcolor='#2ca02c')
    ax2.tick_params(axis='y', labelcolor='#d62728')
    ax.axvspan(1990, 2005, color='grey', alpha=0.15)

fig.suptitle('Case Study: Income Kept Rising, but the HIV/AIDS Epidemic (grey band, ~1990-2005) Crushed Life Expectancy', y=1.03)
plt.tight_layout()
plt.savefig('../Presentation/viz2_case_study_botswana_eswatini.png', dpi=150, bbox_inches='tight')
plt.show()
```

**Visualization 2 Interpretation:**
> In both Botswana and Eswatini, GNI per capita (green) climbs steadily for the full 60-year window with barely a pause. Life expectancy (red), by contrast, rises through the 1980s, then collapses by roughly 20-25 years during the HIV/AIDS epidemic (shaded band), only recovering in the 2000s-2010s as antiretroviral treatment became available. This is the clearest possible illustration of "hollow growth" — a health catastrophe that GNI figures alone completely fail to reveal.

```python
# Visualization 3: Ranked bar chart of the shortlist, ordered by severity of the residual
fig, ax = plt.subplots(figsize=(9, 6))
plot_data = hollow_growth_shortlist.head(10).sort_values('life_gain_residual', ascending=True)
bars = ax.barh(plot_data.index, plot_data['life_gain_residual'], color='#d62728')
ax.set_xlabel('Life expectancy gain vs. expected (years) — more negative = more "hollow"')
ax.set_title('Top 10 Hollow Growth Candidates for the $100M Intervention Fund')
ax.axvline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('../Presentation/viz3_top10_shortlist.png', dpi=150)
plt.show()
```

**Visualization 3 Interpretation:**
> This ranks the ten countries where economic growth diverged most sharply from expected health gains. Botswana and Eswatini stand out as extreme outliers, driven by the HIV/AIDS epidemic documented above; the remaining countries (Guyana, Montenegro, Grenada, Romania, and others) show smaller but still meaningful gaps that merit deeper country-level review before the Committee finalizes fund allocation.

## 5. Summary of Findings & Actionable Recommendations

### 5.1 Key Insights (Summary of Findings)
* Across 1960-2020, GNI-per-capita growth and life-expectancy gains are only weakly correlated once the analysis controls for each country's 1960 starting level — economic growth is not, by itself, a reliable predictor of health improvement.
* A naive comparison (GNI growth vs. raw life expectancy gain) misleadingly flags countries like Ireland, Estonia, and Poland as "hollow growth," when in fact they simply started 1960 with less room to grow (a ceiling effect). Controlling for starting life expectancy is essential to avoid penalizing already-healthy nations.
* Once that ceiling effect is controlled for, Botswana and Eswatini emerge as the clearest, most defensible hollow-growth cases: strong, sustained GNI-per-capita growth throughout 1960-2020, but life expectancy gains roughly 9-10 years short of what their 1960 starting point would predict — driven by the well-documented HIV/AIDS epidemic of the late 1980s-2000s.
* Rapid population growth alone (e.g., Gulf states like Qatar and the UAE) is largely a symptom of labor migration into oil economies, not a sign of "wealth dilution" — these countries' GNI per capita kept climbing right alongside population, so the dilution effect we hypothesized in the problem statement was not the dominant story in the data; the health-lag effect (Section 4.4) was the stronger and more actionable pattern.

### 5.2 Actionable Recommendations
* Prioritize the $100M fund toward the shortlist in Section 4.4 (headed by Botswana and Eswatini), with allocation weighted toward health-system strengthening rather than general economic development, since income growth is not the binding constraint in these countries.
* Before finalizing allocations, commission country-level reviews for the remaining shortlisted nations (e.g. Guyana, Montenegro, Grenada) to confirm the specific driver behind their gap — this index flags a symptom, not a root cause, and different countries likely need different interventions.
* Treat "hollow growth" as a moving target: re-run this analysis periodically (e.g., every 5 years) using the latest data, since new shocks (pandemics, conflict, climate disasters) can create new hollow-growth cases at any time.

### 5.3 Limitations & Areas for Further Research
* GNI per capita and life expectancy are national averages and can mask significant within-country inequality (e.g., a wealthy capital city alongside a struggling rural region).
* The 60-year CAGR / starting-point regression approach is a reasonable first pass, but does not account for non-linear trajectories — a country that dipped badly mid-period and then fully recovered by 2020 could look identical to one that grew smoothly throughout.
* This analysis used population, GNI, and life expectancy only. Incorporating additional Gapminder or World Bank indicators (e.g. under-5 mortality, healthcare spending as % of GDP, education) would sharpen root-cause diagnosis considerably.
* Country name mismatches meant a handful of small states/territories (Monaco, Taiwan, Holy See, etc.) were dropped from the merged panel; this is a minor loss given the fund's likely focus on larger developing nations, but should be noted as a scope limitation.
